```python
datos2["CVE_ENT"] = datos2["Clave_Ent"].astype(str).str.zfill(2)
datos2[["Cve. Municipio"]] = datos2["Cve. Municipio"].astype(str)
datos2.rename(columns={"Cve. Municipio":"CVE_MUN"}, inplace=True)
datos2.drop(["Año", "Clave_Ent"], axis=1, inplace=True)
datos2["Total"] = datos2.sum(axis=1)
```

```python
crime_df = datos2[['CVE_ENT', 'CVE_MUN','Entidad','Municipio','Tipo de delito', 'Subtipo de delito','Total']]
```

```python
crime_df
```

```python
specific_states = crime_df.loc[(crime_df['Entidad'] == 'Hidalgo') |
             (crime_df['Entidad'] =='México') | 
             (crime_df['Entidad'] =='Morelos') | 
            (crime_df['Entidad'] =='Puebla') |
            (crime_df['Entidad'] =='Tlaxcala')]
```

```python
specific_states["Tipo de delito"].value_counts()
```

```python
def agg_crimes(delito):
    df = specific_states.loc[specific_states["Tipo de delito"]==delito]
    df = df.groupby("Municipio").agg({"Total":"sum"})
    df = df.rename(columns={"Total":delito}).reset_index()
    df = df.sort_values(by=delito, ascending=False)
    return df
```

```python
robos_df = agg_crimes("Robo")
homicidios_df = agg_crimes("Homicidio")
secuestro_df = agg_crimes("Secuestro")
feminicidio_df = agg_crimes("Feminicidio")
abuso_sexual_df = agg_crimes("Abuso sexual")
lesiones_df = agg_crimes("Lesiones")
```

## Propiedades

```python
propiedades_df = pd.read_csv("PropiedadesMegalopolis.csv")
propiedades_df.head()
```

```python
propiedades_df["Ciudad"].unique()
```

```python
propiedades_df = propiedades_df.replace("Estado De México", "México")
propiedades_df =propiedades_df.replace("Distrito Federal", "Ciudad de México" )
```

```python
propiedades_df = propiedades_df.loc[(propiedades_df["Superficie total"].str.startswith('Superficie total')) & 
                                   (propiedades_df["Superficie construida"].str.startswith('Superficie construida'))]
```

```python
propiedades_df["Precio"] = propiedades_df["Precio"].str.replace(',','')
propiedades_df["Superficie total"] = propiedades_df["Superficie total"].str.replace(' m²','')
propiedades_df["Superficie total"] = propiedades_df["Superficie total"].str.replace('Superficie total','')
propiedades_df["Superficie construida"] = propiedades_df["Superficie construida"].str.replace(' m²','')
propiedades_df["Superficie construida"] = propiedades_df["Superficie construida"].str.replace('Superficie construida','')
```

```python
propiedades_df["Precio"] = propiedades_df["Precio"].astype(int)
propiedades_df["Superficie total"] = propiedades_df["Superficie total"].astype(float).astype(int)
propiedades_df["Superficie construida"] = propiedades_df["Superficie construida"].astype(float).astype(int)
```

```python
len(propiedades_df["Municipio"].unique())
```

```python
propiedades_df["$/m²"] = propiedades_df["Precio"]/propiedades_df["Superficie total"]
propiedades_df["$/m²(sup_contr)"] = propiedades_df["Precio"]/propiedades_df["Superficie construida"]
```

```python
aggregate_df = propiedades_df.groupby("Municipio").agg({"Precio":"mean", "$/m²":"mean", "$/m²(sup_contr)":"mean"}).reset_index()
aggregate_df
```
