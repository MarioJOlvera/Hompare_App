```python
for x in range(0,len(CasasList)):
    try:
        url = f'https://century21mexico.com{CasasList[x]}'
        Century_request = requests.get(url)
        Century = bs(Century_request.text, "html.parser")
        #time.sleep(4)
        data2 = Century.find('script', type='application/ld+json')
        data2_1 = str(data2)
        data2_1 = data2_1.replace('<script type="application/ld+json">\n', '').strip()
        data2_1 = data2_1.replace('\n', '').strip()
        data2_1 = data2_1.replace('\t', '').strip()
        data2_1 = data2_1.replace('</script>', '').strip()
        data3 = json.loads(data2_1)
        try:
            Coord2 = Century.find('img', class_="img-static-map").get('onclick')
            Coord3 = Coord2.replace('myNavFunc(', '').strip()
            Coord4 = Coord3.replace(')', '').strip()
        except:
            Coord4 = Century.find('img', class_="img-static-map")
        DataT = Century.find('div', class_="col-lg-6").text
        DataT = DataT.replace('\n', '').strip()
        Tipo = re.findall('Tipo:(.+)Precio de venta:', DataT)
        Precio = re.findall('Precio de venta:(.+)Superficie de terreno:', DataT)
        Terreno = re.findall('Superficie de terreno:(.+)Superficie de construccin:', DataT)
        Construccion = re.findall('Superficie de construccin:(.+)Superficie de frente:', DataT)
        Recamaras = re.findall('Recmaras:(.+)Baos completos:', DataT)
        BanosC = re.findall('Baos completos:(.+)Medios Baos:', DataT)
        MBanos = re.findall('Medios Baos:(.+)Estacionamientos:', DataT)
        Estacionamientos = re.findall('Estacionamientos:(.+)Edad:', DataT)
        Niveles = re.findall('Niveles construidos:(.+)Cocina:', DataT)
        #print(f"OK {CasasList[x]} corresponde a {x}")
        #Propiedades["Categoria"].append(Valores[0].getText().replace('\n', '').strip())
        Propiedades["Tipo"].append(data3['@type'])
        #print("OK Tipo")
        #Propiedades["V o R"].append()
        Propiedades["Pais"].append(data3['address']['addressCountry'])
        #print("OK Pais")
        Propiedades["Ciudad"].append(data3['address']['addressRegion'])
        #print("OK Ciudad")
        Propiedades["Municipio"].append(data3['address']['addressLocality'])
        #print("OK Municipio")
        Propiedades["Calle"].append(data3['address']['streetAddress'])
        #print("OK Calle")
        Propiedades["Codigo_Postal"].append(data3['address']['postalCode'])
        #print("OK Codigo Postal")
        Propiedades["Precio"].append(data3['potentialAction']['priceSpecification']['price'])
        #print("OK Precio")
        Propiedades["Moneda"].append(data3['potentialAction']['priceSpecification']['priceCurrency'])
        #print("OK Moneda")
        #Propiedades["Superficie_Total"].append()
        Propiedades["Superficie_Construida"].append(data3['floorSize'])
        #print("OK Sup Construida")
        Propiedades["Recamaras"].append(data3['numberOfRooms'])
        #print("OK Recamaras 1")
        #Propiedades["Banos"].append()
        #Propiedades["Estacionamientos"].append()
        Propiedades["DataT"].append(DataT)
        #print("OK DataT")
        Propiedades["Tipo2"].append(Tipo)
        #print("OK Tipo2")
        Propiedades["Precio2"].append(Precio)
        #print("OK Precio2")
        Propiedades["Superficie_Total2"].append(Terreno)
        #print("OK Terreno")
        Propiedades["Superficie_Construida2"].append(Construccion)
        #print("OK Construccion")
        Propiedades["Recamaras2"].append(Recamaras)
        #print("OK Recamaras")
        Propiedades["Banos2"].append(BanosC)
        #print("OK BanosC")
        Propiedades["MBanos"].append(MBanos)
        #print("OK MBanos")
        Propiedades["Estacionamientos2"].append(Estacionamientos)
        #print("OK Estacionamientos2")
        Propiedades["Niveles"].append(Niveles)
        #print("OK Niveles")
        Propiedades["CordA"].append(Coord4)
        #print("OK CordA")
        Propiedades["URL"].append(f'https://century21mexico.com{CasasList[x]}')
        print(f"OK {x} de {len(CasasList)}")
        
    except:
        print(f"Error Pagina {x}")
        pass

Propiedades_DF = pd.DataFrame(Propiedades)
Propiedades_DF.to_csv('Propiedades.csv', index=False, encoding='utf-8-sig')
Propiedades_DF
```
