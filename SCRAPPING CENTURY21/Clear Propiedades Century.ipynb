```python
#Import CSV of Scrapping from 1M to 1.25M

PropiedadesMegalopolis = "PropiedadesMegalopolis.csv"


# Read CSVs and store into Pandas data frames
PropiedadesMegalopolis_DF = pd.read_csv(PropiedadesMegalopolis)
PropiedadesMegalopolis_DF = PropiedadesMegalopolis_DF.drop(columns=['Categoria','c_estado.1', 'c_tipo_asenta.1', 'c_mnpio.1','Unnamed: 20', 'Tipo', 'V o R', 'Propiedad','Coordenadas'])
PropiedadesMegalopolis_DF['Precio'] = PropiedadesMegalopolis_DF['Precio'].str.replace(',','')
PropiedadesMegalopolis_DF = PropiedadesMegalopolis_DF.astype({"Precio": int})
PropiedadesMegalopolis_DF["Superficie total"] = PropiedadesMegalopolis_DF["Superficie total"].str.replace("Superficie total", "")
PropiedadesMegalopolis_DF["Superficie total"] = PropiedadesMegalopolis_DF["Superficie total"].str.replace(" m²", "")
PropiedadesMegalopolis_DF["Superficie construida"] = PropiedadesMegalopolis_DF["Superficie construida"].str.replace("Superficie construida", "")
PropiedadesMegalopolis_DF["Superficie construida"] = PropiedadesMegalopolis_DF["Superficie construida"].str.replace(" m²", "")
PropiedadesMegalopolis_DF["Estacionamientos"] = PropiedadesMegalopolis_DF["Estacionamientos"].str.replace("Estacionamientos", "")
PropiedadesMegalopolis_DF["Recámaras"] = PropiedadesMegalopolis_DF["Recámaras"].str.replace("Recámaras", "")
PropiedadesMegalopolis_DF["Baños"] = PropiedadesMegalopolis_DF["Baños"].str.replace("Baños", "")
PropiedadesMegalopolis_DF = PropiedadesMegalopolis_DF.astype({"Superficie total": float})
PropiedadesMegalopolis_DF = PropiedadesMegalopolis_DF.astype({"Superficie construida": float})
PropiedadesMegalopolis_DF = PropiedadesMegalopolis_DF.astype({"Baños": float})
PropiedadesMegalopolis_DF = PropiedadesMegalopolis_DF.astype({"Recámaras": float})
PropiedadesMegalopolis_DF = PropiedadesMegalopolis_DF.astype({"Estacionamientos": float})
PropiedadesMegalopolis_DF.round(0) 
PropiedadesMegalopolis_DF


```

```python
PropiedadesMegalopolis_DF.rename(columns={'Superficie total':'Superficie_Total',
                         'Superficie construida':'Superficie_Construida',
                        }, 
                 inplace=True)
PropiedadesMegalopolis_DF
```

```python
#PropiedadesMegalopolis_DF.groupby(['c_mnpio']).mean()
PromXMun = PropiedadesMegalopolis_DF.groupby("c_mnpio").agg(
    min_Price=('Precio', 'min'),
    max_Price=('Precio', 'max'),
    Conteo_inmbuebles=('Precio', 'count'),
    Promedio_Precio = ('Precio', 'mean'),
    Promedio_Terreno = ('Superficie_Total', 'mean'),
    Promedio_Construccion = ('Superficie_Construida', 'mean'),
    Promedio_Recamaras = ('Recámaras', 'mean'),
    Promedio_Baños = ('Baños', 'mean'),
)
PromXMun
```

```python
PromXMun.round({"Promedio_Precio":2, "Promedio_Terreno":2, "Promedio_Construccion":2,
                "Promedio_Recamaras":2, "Promedio_Baños":2}) 
PromXMun = PromXMun.astype({"Promedio_Baños": int})
PromXMun = PromXMun.astype({"Promedio_Recamaras": int})
PromXMun = PromXMun.astype({"Promedio_Construccion": int})
PromXMun = PromXMun.astype({"Promedio_Terreno": int})
PromXMun = PromXMun.astype({"Promedio_Precio": int})
PromXMun
```
